Resizing the image using pillow library

```python
from PIL import Image # importing an image  module from pillow library
```

```python
img = Image.open('/content/cat_img.jpg')
resized_img=img.resize((200,200)) # converting an image from 1200x1600 to 200x200 sized
```

```python
resized_img.save('resized_cat_img.jpg')
```

```python
# displaying the image from numpy array
cat_img_res=mpimg.imread('/content/resized_cat_img.jpg')
img_res_plot = plt.imshow(cat_img_res)
```

```python
print(cat_img_res.shape)
```

Converting an RGB image into a Gray scale image

```python
# Imporing an OpenCV library
import cv2
```

```python
img_cv=cv2.imread('/content/cat_img.jpg')
```

```python
type(img_cv)
```

```python
img_cv.shape
```

```python
gray_image=cv2.cvtColor(img_cv,cv2.COLOR_RGB2GRAY)
```

```python
gray_image_plot=plt.imshow(gray_image)
gray_image.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image
cv2_imshow(gray_image)
```

```python
# saving the grayscale image
cv2.imwrite('cat_grayscale_image.jpg', gray_image)
```
